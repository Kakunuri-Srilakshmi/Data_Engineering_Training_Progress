## Task - 1 - Sales Data Visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Prepare Dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Revenue": [12000, 15000, 17000, 14000, 18000, 22000,
                24000, 20000, 23000, 25000, 27000, 30000],
    "Profit": [3000, 3500, 4000, 3200, 4500, 6000,
               6500, 5800, 6200, 7000, 7500, 8000],
    "Region": ["North", "South", "East", "West",
               "North", "South", "East", "West",
               "North", "South", "East", "West"]
}
df = pd.DataFrame(data)
```

## Line Chart


```python
# 1. Line Chart - Revenue over Months
plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Revenue"], marker="o", linestyle="-", color="b")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

## Bar Chart


```python
# 2. Bar Chart - Profit for each Month
plt.figure(figsize=(8,5))
plt.bar(df["Month"], df["Profit"], color="orange")
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

## Grouped Bar Chart

```python
# 3. Grouped Bar Chart - Revenue vs Profit
x = range(len(df["Month"]))
plt.figure(figsize=(10,6))
plt.bar(x, df["Revenue"], width=0.4, label="Revenue")
plt.bar([i+0.4 for i in x], df["Profit"], width=0.4, label="Profit")
plt.xticks([i+0.2 for i in x], df["Month"])
plt.title("Revenue vs Profit per Month")
plt.xlabel("Month")
plt.ylabel("Value")
plt.legend()
plt.show()
```

## Pie Chart

```python
# 4. Pie Chart - Revenue contribution by Region
region_revenue = df.groupby("Region")["Revenue"].sum()
plt.figure(figsize=(6,6))
plt.pie(region_revenue, labels=region_revenue.index, autopct="%1.1f%%", startangle=90)
plt.title("Revenue Contribution by Region")
plt.show()
```

## Scatter Plot

```python
# 5. Scatter Plot - Revenue vs Profit
plt.figure(figsize=(7,5))
plt.scatter(df["Revenue"], df["Profit"], color="green")
plt.title("Revenue vs Profit")
plt.xlabel("Revenue")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```

## Histogram

```python
# 6. Histogram - Distribution of Revenue
plt.figure(figsize=(8,5))
plt.hist(df["Revenue"], bins=6, color="purple", edgecolor="black")
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

## Box Plot

```python
# 7. Box Plot - Profit across Regions 
plt.figure(figsize=(8,5))
sns.boxplot(x="Region", y="Profit", data=df, palette="Set2")
plt.title("Profit Distribution across Regions")
plt.show()
```
